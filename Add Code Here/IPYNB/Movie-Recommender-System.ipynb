```python
#since we require only the first 3 names of the people from cast,
# As they are the mostly the main characters in the movies, and apporopriate enough to consider for our recommender system

def convert_cast(obj):                    
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
            
        else:
            break
    return L
         
```

#### Now in our tags column we can see that many words are similar, for example : action, actions is similar ; actor, actors is similar. Now we want that there should be only one word representing all the similar words. So we'll use stem method in scikit learn to do that

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#for example

ps.stem('dancing')
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)

#defining a function named stem to apply the stem feature to all rows
```

```python
new_movies['tags'].apply(stem)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

#### Now to calculate the similarity between movies we'll calculate this similarity using a similarity measure function known as cosine similarity

#### 1 - Higher Similarity between movies  ;  0 - Lesser Similarity between movies

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(tags_vector)   

#1st movie's cosine distance with 1st movie (i.e itself) will be 1.0 ; as we can see in table
#similarly the distance is calculated from one movie to every another movie in the given dataset, that means...
# one movie's distance is calculated with respect to 4806 movies present in the dataset.
```

```python
movie_similarity = cosine_similarity(tags_vector) 
```

#### making a recommend function to recommend 5 similar movies based on the input provided.

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]   #extracting the movie's index here
    distances = movie_similarity[movie_index]   #extracting the movie similarity score and storing it in distances variable, 
    
    #basically sorting the movie_similarity array in descending order, so that we can get the closest values.
    
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(i[0],new_movies.iloc[i[0]].title)
 
```

```python
recommend('Avatar')
```

```python
recommend('Batman')  #recommending similar movies same as the Batman
```
