```python
#=================================================================================================
#DESCRIÇÃO: Algoritmo proposto para o calculo dos resultados matriciais feito
#           pelo método da fatoração LU
#=================================================================================================
import numpy as np

#=================================================================================================
def fatoraLU(A):
    #Função para fazer uma copia da matriz A.
    U = np.copy(A)
    
    #Função shape serve para remodular uma matriz atribuindo a uma tupla de dimensões.
    n = np.shape(U)[0]
    
    #Retorna um array com uns na diagonal e zeros em outro lugar.
    L = np.eye(n)
    
    #Função arange retorna valores uniformemente espaçados dentro de um determinado intervalo.
    for j in np.arange(n-1):  
        for i in np.arange(j+1,n):  
            L[i,j] = U[i,j]/U[j,j]  
            for k in np.arange(j+1,n):  
                U[i,k] = U[i,k] - L[i,j]*U[j,k]  
            U[i,j] = 0  
            
    return L, U
#=================================================================================================

#=================================================================================================
#Exemplo de uma matriz qualquer
#=================================================================================================
A = [[1.0,1.0,0,3.0],
     [2.0,1.0,-1.0,1.0],
     [3.0,-1.0,-1.0,2.0],
     [-1.0,2.0,3.0,-1.0]]
#=================================================================================================
#=================================================================================================
#Valores para uma possivel solução 
#=================================================================================================
b = [4.0, 1.0, -3.0, 4.0]
#=================================================================================================

#=================================================================================================
#Chamada da função para o calculo dos valores dos coeficientes
#=================================================================================================
y,x = fatoraLU(A)
#=================================================================================================

#=================================================================================================
#Função ligalg.solve retorna a solução dos coeficientes da matriz.
#=================================================================================================
sol_y = np.linalg.solve(y,b)    
sol_x = np.linalg.solve(x,sol_y)

#=================================================================================================

#=================================================================================================
#Amostragem dos Resultados Obtidos
#=================================================================================================
print("\nMatriz A")
print(A)
print("\nMatriz L")
print(y)
print("\nMatriz U")
print(x)
print("\n Soluções para y")
print(sol_y)
print("\nSoluções para X")
print(sol_x)
#=================================================================================================
```
